## CONSITENCY

```python
for measure in range(3):
    print('measure:',measure)
    tau,p=stats.kendalltau(story_pretrain[:,:,measure], review_pretrain[:,:,measure])
    print(tau,p)
```

```python
for measure in range(3):
    print('measure:',measure)
    taus=[]
    for i in range(10):
        tau,p=stats.kendalltau(story_neg[i,:,:,measure], review_neg[i,:,:,measure])
        print(tau,p)
        taus.append(tau)
    print('\n')
    print(np.average(taus),np.std(taus))
    print('\n')
```

```python
for measure in range(3):
    print('measure:',measure)
    taus=[]
    for i in range(10):
        tau, p = stats.kendalltau(story_control[i,:,:,measure], review_control[i,:,:,measure])
        print(tau,p)
        taus.append(tau)
    print('\n')
    print(np.average(taus),np.std(taus))
    print('\n')
```

## richer get richer


```python
for measure in range(3):
    print('measure:',measure)
    taus=[]
    for i in range(10):
        tau,p=stats.kendalltau(story_pretrain[:,:,measure], story_neg[i,:,:,measure]-story_pretrain[:,:,measure])
        print(tau,p<0.05)
        taus.append(tau)
    print(np.average(taus),np.std(taus))
```
